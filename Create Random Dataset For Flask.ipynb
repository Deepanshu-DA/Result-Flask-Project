```python
import pandas as pd
import numpy as np
from faker import Faker
```

```python
fake = Faker()
random_data = [(fake.name(), fake.random_element(['Male', 'Female'])) for _ in range(10000)]
```

```python
df = pd.DataFrame(random_data, columns=['Name', 'Gender'])
```

```python
df.insert(0, column = 'Student_Roll_no', value = (np.arange(1,10001,1)))
```

```python
start_date = '2008-01-01'
end_date = '2009-12-31'
num_dates = len(df)

dob = pd.to_datetime(np.random.choice(pd.date_range(start_date, end_date), num_dates))
df.insert(3, column = "DOB", value = (dob))
```

```python
df.insert(4, column = "Father's_Name", value = ([(fake.name_male())for _ in range(10000)]))
```

```python
df.insert(5, column = "Mother's_Name", value = ([(fake.name_female())for _ in range(10000)]))
```

```python
def Code(id):
    if 9001 <=id <= 10000:
        return 5100
    elif 8001 <= id <= 9000:
        return 5090
    elif 7001 <= id <= 8000:
        return 5080
    elif 6001 <= id <= 7000:
        return 5070
    elif 5001 <= id <= 6000:
        return 5060
    elif 4001 <= id <= 5000:
        return 5050
    elif 3001 <= id <= 4000:
        return 5040
    elif 2001 <= id <= 3000:
        return 5030
    elif 1001 <= id <= 2000:
        return 5020
    elif 1 <= id <= 1000:
        return 5010
    else:
        return 'error'
df['School_code'] = df['Student_Roll_no.'].apply(Code)
```

```python
df.insert(7, column = "English", value = (np.random.randint(10,100,10000)))
```

```python
def assign_grade(marks):
    if 91 <= marks <= 100:
        return 'A1'
    elif 81 <= marks <= 90:
        return 'A2'
    elif 71 <= marks <= 80:
        return 'B1'
    elif 61 <= marks <= 70:
        return 'B2'
    elif 51 <= marks <= 60:
        return 'C1'
    elif 41 <= marks <= 50:
        return 'C2'
    elif 31 <= marks <= 40:
        return 'D'
    elif 21 <= marks <= 30:
        return 'E1'
    else:
        return 'E2'
```

```python
df['English_Grade'] = df['English'].apply(assign_grade)
```
